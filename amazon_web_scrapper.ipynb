```python
# Import Libraries
from bs4 import BeautifulSoup
import csv
from msedge.selenium_tools import Edge, EdgeOptions 
import time
import pandas as pd
```

```python
# Installing the driver
options = EdgeOptions()
options.use_chromium = True
driver =  Edge(options = options)
```

```python
#getting url list form csv
with open('Price_Tracker.csv', encoding= "utf-8-sig") as csvfile:
    reader = csv.DictReader(csvfile)
    count = 0
    urls=[]
    for row in reader:
        count = count + 1
        urls.append(row["search"])
```

```python
# Accessing the URL
for url in urls:
 driver.get(url)
 # Creating a soup
 soup = BeautifulSoup(driver.page_source, 'html.parser')
 title = soup.find(id = 'productTitle').get_text()
 stars = soup.find('span' , { 'data-hook':'rating-out-of-text'})
 if stars is None:
    stars = 'Not Available'
 else:
        stars = stars.get_text()
 date = time.ctime()
 ratings = soup.find( id = 'acrCustomerReviewText' )
 if ratings is None:
    ratings = 'Not Available'
 else:
        ratings = ratings.get_text()

 title = title.strip()
 ratings = ratings.strip()
 stars = stars.strip()

 data = [title,ratings,stars,url,date]
# adding more rows
 with open('Tracker.csv','a+', newline='',encoding='utf-8') as f:    
    writer = csv.writer(f)
    writer.writerow(data)
```
